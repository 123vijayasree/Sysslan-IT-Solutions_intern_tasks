```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned dataset
df = pd.read_csv("/content/cleaned_data.csv")
sns.set(style="whitegrid")
```

```python
#Task 5.1: Pivot Table (Station-wise Distribution)

pivot_table = pd.pivot_table(
    df,
    values='Train_No',
    index='Station_Name',
    aggfunc='count'
).rename(columns={'Train_No': 'Train_Count'})

print("\nPivot Table: Station-wise Train Distribution")
print(pivot_table.head())
```

```python
# Task 5.2: Cross-tab (Station vs Route Type)
cross_tab = pd.crosstab(
    df['Station_Name'],
    df['Route_Number']
)

print("\nCross-tab: Station vs Route Type")
print(cross_tab.head())
```

```python
# Task 5.3: Visualizations

# 1. Pivot Table Visualization
plt.figure()
pivot_table.sort_values(by='Train_Count', ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Stations by Train Count")
plt.xlabel("Station")
plt.ylabel("Number of Trains")
plt.show()

# 2. Cross-tab Visualization (Stacked Bar)
plt.figure()
cross_tab.head(10).plot(kind='bar', stacked=True)
plt.title("Station vs Route Type Distribution")
plt.xlabel("Station")
plt.ylabel("Train Count")
plt.show()
```

```python
# Task 5.4: Advanced Insight
print("\nAdvanced Insights:")

# Busiest station
busiest_station = pivot_table['Train_Count'].idxmax()
print(f"- {busiest_station} is the busiest station.")
# Route type dominance
route_summary = df['Route_Number'].value_counts()
print("- Route distribution:")
print(route_summary)
# Observation message
print("- Some stations handle multiple route types (Short, Medium, Long).")
print("- High-traffic stations are key transport hubs.")
```
