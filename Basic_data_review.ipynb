```python
import pandas as pd
import numpy as np
```

```python
#Task 1.1: Overview of dataset
df = pd.read_csv("/content/Dataset1.csv")

print("Shape:", df.shape)
print("Columns:", df.columns)
```

```python
#Task 1.2: List all trains with source & destination

source_stations = df.loc[df.groupby('Train_No')['Distance'].idxmin()][['Train_No', 'Station_Name']]
source_stations = source_stations.rename(columns={'Station_Name': 'Source_Station'})

destination_stations = df.loc[df.groupby('Train_No')['Distance'].idxmax()][['Train_No', 'Station_Name']]
destination_stations = destination_stations.rename(columns={'Station_Name': 'Destination_Station'})

train_routes = pd.merge(source_stations, destination_stations, on='Train_No', how='inner')

# Display unique train routes
train_routes.drop_duplicates().reset_index(drop=True)
```

```python
#Task 1.3: Number of stops per train
stops = df.groupby('Train_No')['Station_Name'].count()
print(stops)
```

```python
#Task 1.4: Max & Min stops train
print("Max stops train:", stops.idxmax())
print("Min stops train:", stops.idxmin())
```
