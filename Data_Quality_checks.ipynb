```python
import pandas as pd
import numpy as np
```

```python
#Load dataset
df = pd.read_csv("/content/Dataset1.csv")
```

```python
#Task 3.1: Handle missing values
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.fillna(method='ffill', inplace=True)
```

```python
#Task 3.2: Remove duplicates
df.drop_duplicates(inplace=True)
print("Before:", len(df))
df.drop_duplicates(inplace=True)
print("After:", len(df))
```

```python
#Task 3.3: Verify station order
df.sort_values(by=['Train_No','SN'], inplace=True)
df.sort_values(by=['Train_No','Arrival_time'], inplace=True)
```

```python
df_sorted = df.sort_values(by=['Train_No','SN'])

print(df_sorted[['Train_No','Station_Name','SN']].head(10))
```

```python
#Task 3.4: Save cleaned dataset
df.to_csv("cleaned_data.csv", index=False)
print("Cleaned dataset saved successfully!")
```
