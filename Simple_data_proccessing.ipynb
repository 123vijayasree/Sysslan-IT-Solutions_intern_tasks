```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Dataset1.csv")
```

```python
#Task 2.1: Standardize time (Arrival & Departure)
df['Arrival_time'] = pd.to_datetime(df['Arrival_time'], errors='coerce')
df['Departure_Time'] = pd.to_datetime(df['Departure_Time'], errors='coerce')

print(df[['Arrival_time','Departure_Time']].head())
```

```python
#Task 2.2: Compute journey duration
df['Duration'] = df['Departure_Time'] - df['Arrival_time']

print(df[['Train_No','Duration']].head())
```

```python
# Task 2.3: Classify routes (Short / Medium / Long)
df['Duration_hours'] = df['Duration'].dt.total_seconds() / 3600
```

```python
def classify_route(x):
    if x < 2:
        return "Short"
    elif x < 5:
        return "Medium"
    else:
        return "Long"

df['Route_Type'] = df['Duration_hours'].apply(classify_route)

print(df[['Train_No','Route_Type']].head())
```
